```python
import numpy as np
```

```python
# load numpy array from npy file
from numpy import load
# load array
class_loss = np.load('loss_class_epoch15_1.npy')
regressor_loss = np.load('loss_regr_epoch15_1.npy')
total_loss = np.load('loss_tot_epoch15_1.npy')

val_class_loss = np.load('val_loss_class_epoch15_1.npy')
val_regressor_loss = np.load('val_loss_regr_epoch15_1.npy')
val_total_loss = np.load('val_loss_tot_epoch15_1.npy')

# print the array
print(class_loss.shape)
print(regressor_loss.shape)
print(total_loss.shape)
print(class_loss[0])
print (class_loss)
display (class_loss)
```

```python
import matplotlib.pyplot as plt 

plt.plot(class_loss)
plt.title('Training Class Loss')
plt.xlabel('per ten updates')
plt.ylabel('train error')
plt.show()

plt.plot(regressor_loss)
plt.title('Training Regressor Loss')
plt.xlabel('per ten updates')
plt.ylabel('train error')
plt.show()

plt.plot(total_loss)
plt.title('Training Total Loss')
plt.xlabel('per ten updates')
plt.ylabel('train error')
plt.show()


#plt.plot(val_class_loss)
plt.plot(range(0, int(len(class_loss)/len(val_class_loss))*len(val_class_loss), int(len(class_loss)/len(val_class_loss))),val_class_loss)
plt.title('Validation Class Loss')
plt.xlabel('per ten updates')
plt.ylabel('validation error')
plt.show()

#plt.plot(val_regressor_loss)
plt.plot(range(0, int(len(class_loss)/len(val_class_loss))*len(val_regressor_loss), int(len(regressor_loss)/len(val_regressor_loss))),val_regressor_loss)
plt.title('Validation Regressor Loss')
plt.xlabel('per ten updates')
plt.ylabel('validation error')
plt.show()

#plt.plot(val_total_loss)
plt.plot(range(0, int(len(total_loss)/len(val_total_loss))*len(val_total_loss), int(len(total_loss)/len(val_total_loss))),val_total_loss)
plt.title('Validation Total Loss')
plt.xlabel('per ten updates')
plt.ylabel('validation error')
plt.show()
```
